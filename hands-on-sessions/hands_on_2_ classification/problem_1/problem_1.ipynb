##**Data acquisition**

```python
import pandas as pd
```

```python
df = pd.read_csv("ionosphere.data")
```

```python
print(df.shape)
```

```python
df
```

```python
# rename the columns of the dataframe using the atribute "columns" of the dataframe object
columns = list() #[]
for i in range(1,18):
  col1 = "f"+str(i)+"_r"
  col2 = "f"+str(i)+"_i"
  columns.append(col1)
  columns.append(col2)

columns.append("target")
df.columns = columns
df.head()
```

##**Pre processing data**

```python
# categorize the target in 0 and 1 from the labels "b" and "g"
targetCategory = {"g": 1, "b":0}
df["target"] = df["target"].map(targetCategory)
```

```python
df["target"]
```

##**Feature engineering**

```python
# analize the balance of the data
dataBalanced = df["target"].value_counts()
print(dataBalanced)
```

```python
# balance the data removing rows of the class with the majority
counter = 98
dfAux = pd.DataFrame()
for index,row in df.iterrows():
  if(counter>0):
    if(row["target"]==1):
      counter-=1
    else:
        dfAux = dfAux.append(row)
  else:
    dfAux = dfAux.append(row)

```

```python
dfAux.shape
```

```python
dataBalanced = dfAux["target"].value_counts()
print(dataBalanced)
```

##**Data preparation**

```python
features = dfAux.loc[:, dfAux.columns != "target"]
target = dfAux.loc[:, dfAux.columns == "target"]
```

```python
print(features)
print(target)
```

```python
# analisis para escalamiento de datos si es que es necesario
features.min()
```

##**Modeling**

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2)
print(target_train.shape, target_test.shape)
```

```python
from keras.models import Sequential
from keras.layers.core import Activation, Dense
import tensorflow as tf
```

```python
myInputSize = features_train.shape[1]

# arquitectura de NN
model = Sequential()
model.add(Dense(32, input_dim=myInputSize))
model.add(Activation("tanh"))
model.add(Dense(1))
model.add(Activation("relu"))

model.compile(loss="binary_crossentropy", optimizer="rmsprop", metrics=[tf.keras.metrics.Accuracy()])
```

```python
history = model.fit(features_train, target_train, epochs=20, validation_split=0.2, verbose=2)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
```

##**Classification and validation of the models**

```python
yPred = model.predict(features_test)
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

cm = confusion_matrix(np.round(yPred), target_test, labels=[0.0,1.0])
```
