### **Useful information about structured data type**

#### **Lists:**
- They are sequences of elements, which can be of any type: int, float, string and even other lists.
- **Mutable** and ordered.
- Each element of the list corresponds to a position or index, which can be accessed as with strings.
- Declared with: "[" and "]"

#### **Tuples:**
- They are sequences of elements, which can be of any type: int, float, string and even other tuples.
- **Immutable** and ordered.
- Each element of the tuple corresponds to a position or index, which can be accessed as with strings.
- Declared with: "(" and ")"

#### **Tuples VS Lists:**

- Functions that can be used for both lists and tuples: len(), max(), min(), sum(), any(), all(), sorted()       
- Methods that cannot be used for tuples: append(), insert(), remove(), pop(), clear(), sort(), reverse()



**Another very important structured data of Python are the dictionaries!**

#### **Dictionaries:**
- Dictionaries are groupings of data of type: Key -> Value. 
- They have no order. 
- Declared using "{" and "}". The syntax is: my_dictionary = {key_1: "value", key_2: "value"}

####**Dictionary operations:**
- clear() – Remove all the elements from the dictionary
- copy() – Returns a copy of the dictionary
- get() – Returns the value of specified key
- items() – Returns a list containing a tuple for each key value pair
- keys() – Returns a list containing dictionary’s keys
- pop() – Remove the element with specified key
- popitem() – Removes the last inserted key-value pair
- update() – Updates dictionary with specified key-value pairs
- values() – Returns a list of all the values of dictionary

## **Control structures**

In programming languages, control structures allow you to modify the execution flow of a program's instructions.
All control structures have a single entry point. Control structures can be classified into: 
- **Sequential control structure.**
- **Selection control structure.**
- **Iteration control structure.**


>❗The main specifics structures of the structured programming paradigm. 
[Boehm, Corrado, & Jacopini, Giuseppe (1966), "Flow Diagrams, Turing Machines, and Languages with only Two Formation Rules", Communications of the ACM 9(5): 366-371](https://cse.buffalo.edu/~rapaport/111F04/greatidea3.html)


#### **Selection structures in Python:**

The selection control structure permits the execution of one block of program statements if the condition is true. It permits another block of program statements if the condition is false.

Syntax of a if-else structure:

```
if (condition):
    # Executes this block if
    # condition is true
else:
    # Executes this block if
    # condition is false```



````python
# example
a = 10
b = -10
c = 0
  
if a > 0 or b > 0:
    print("Either of the number is greater than 0")
else:
    print("No number is greater than 0")
  
if b > 0 or c > 0:
    print("Either of the number is greater than 0")
else:
    print("No number is greater than 0")
````

#### **Iterative structures in Python:**

Python provides three ways for executing the loops. While all the ways provide similar basic functionality, they differ in their syntax and condition checking time.
This tutorial will show the two main iterative structures:
- **While**
```
while expression:
    statement(s)
```

- **For**
```
for iterator_var in sequence:
    statements(s)
```



**NOTE:**
The structures in Python need to be writed in [Pythonic notation](https://peps.python.org/pep-0008/). 
Structuring by Indentation is the space between each line of code and the margin. Unlike many other programming languages, indentation in python is not just for aesthetics but defines the structure of the code.

````python
# example of "while"
count = 0
while (count < 3):   
    count = count + 1
    print("inside of a while structure")
````

````python
# example of "for"
count = 0
for i in range(0, count):
    print("inside of a for structure")
````

## **Libraries (or packages) and Frameworks**

In programming, when we talk about a library, we refer to a set of code files that provide us some functionalities. Libraries allow us to reuse code. 

Each python package is made up of one or more Python files (modules).

When we talk about Framework, we refer to a broader concept, not only to implemented functionalities but to a whole framework that can include one or more libraries and tools.

### **First, the first! Exist hundreds packages/modules in Python, so, If we need to use any of it, We need to import it.**

The Python *import* statement lets you import a module into your code. A module is a file that contains functions and values that you can reference from your program. The import statement syntax is: import modulename.

Some import options:


**1- Importing modules:**
```
# syntax

```
- For example:
```

```

**2- Using the "from" Statement:**
```
# syntax
from [module] import [function or value]
```
- For example:
```

```

**Using alias**

We can set an alias to a module. Using an alias, we can call the module and its functions/values, directly using this alias

```
# syntax

```
- For example:
```

####**Slicing data**

Pandas is very useful for the manipulation of mathematical data and is widely used in the field of machine learning. It comprises many methods for its proper functioning. **loc()** and **iloc()** are one of those methods. These are used in slicing data from the Pandas DataFrame. They help in the convenient selection of data from the DataFrame. They are used in filtering the data according to some conditions.

- loc: lhe loc() function **is label based** data selecting method which means that we have to pass the name of the row or column which we want to select. 

- iloc: the iloc() function **is an indexed-based** selecting method which means that we have to pass an integer index in the method to select a specific row/column. 

Differences between **loc** and **iloc**:
- This method includes the last element of the range passed in it, unlike iloc().
- loc() can accept the boolean data unlike iloc().

####**Using loc() function:**

####**Selecting a columns of a DataFrame**

````python
df.loc[:,'Location']
````

####**Selecting data according to some conditions**

````python
display(df.loc[(df['Rocket'] == 'Success')])
````

####**Selecting a range of rows from the DataFrame**

````python
df.loc[2:5,["Location","Rocket"]]
````

####**Updating the value of any column**


````python
display(df.loc[df.Rocket=="65","Rocket"])
df.loc[df.Rocket=="65","Rocket"]=25
display(df.loc[df.Rocket=="25","Rocket"])
````

####**Using iloc() function:**

- Selecting rows using integer indices.
- Selecting a range of columns and rows simultaneously.

####**Selecting rows using integer indices**

````python
df.iloc[[0,2,7,9]]
````

####**Selecting a range of columns and rows simultaneously**

````python
df.iloc[1: 5, 2: 5]
````
